```python
#Khaled's implementation of Part 1 UTXO manager
class UTXOManager:
    def __init__(self):
        # Store UTXOs as dictionary: (tx_id, index)-> (amount, owner)
        self.utxo_set = {}
        #here we have key:value pair where the key is ,a tuple (i.e it's immutable), (tx_id,index) pair. 
        #value itself is another dict having (amount,owner) pair
        # {key :{owner: amount}}
    
    def add_utxo(self, tx_id: str, index: int, amount: float, owner: str):
        """Add a new UTXO to the set."""
        key=(tx_id,index)

        
        #following are some basic checks such as no -ve money and no duplicates.
        if(amount<=0):
            print("UTXO cannot have negative or 0 BTC!!!")
        elif key in self.utxo_set:
            print(f"the damn key : {key} already exists!!!")
        else:
        #if no conditions are violated, then we store the key
            self.utxo_set[key]={"amount":amount,"owner":owner}
        
    def remove_utxo(self, tx_id: str, index: int):
        """Remove a UTXO (when spent)."""
        key=(tx_id,index) #this forms the identifier which we will then search in the dict and kick it out

        #suppose the key doesn't exist. then you cannot remove
        if key not in self.utxo_set:
            print("fools, there was no such key!!!")
        else:
            del self.utxo_set[key]
    
    def get_balance(self, owner: str)-> float:
        """Calculate total balance for an address."""
        #let initial balance be 0
        balance = 0.0
        for values in self.utxo_set.values:
            if(values["owner"]==owner):
                balance+=values["amount"]
        return balance
        
    def exists(self, tx_id: str, index: int)-> bool:
        """Check if UTXO exists and is unspent."""
        #this is obvious
        key=(tx_id,index)
        return key in self.utxo_set
    
    def get_utxos_for_owner(self, owner: str)-> list:
        """Get all UTXOs owned by an address."""
        list=[]
        for (tx_id,index),values in self.utxo_set.items():
            if(values["owner"]==owner):
                list.append({
                    "tx_id":tx_id,
                    "index":index,
                    "owner":owner,
                    "amount":values["amount"]
                })

        return list         
```
